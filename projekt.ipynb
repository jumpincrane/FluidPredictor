# Projekt

## Cel projektu
Na Uniwersytecie Medycznym w Poznaniu przeprowadzono badanie mające sprawdzić wpływ [bruksizmu](https://pl.wikipedia.org/wiki/Bruksizm) i innych wskaźników na objętość płynu dziąsłowego. 

Program ma zwracać wartość reprezentującą objętość płynu dziąsłowego dla konkretnych zębów na podstawie przeprowadzonych pomiarów i wywiadu z pacjentem.

## Dane wyjściowe
U każdego uczestnika eksperymentu pobrano płyn dziąsłowy z 12 szczelin dziąsłowych - przy 6 zębach - w każdym zębie od strony policzka (**B**uccal) i podniebienia (**P**alatal).
Czyli B od zewnątrz, P od środka.

Numery zębów to są najpierw kwadranty jamy ustnej, potem numer zęba, np. 16-B to pierwszy kwadrant, szóstka od zewnątrz.

mouth.png




## Dane wejściowe


*   API (Approximal Plaque Index) - stosunek przestrzeni międzyzębowych z płytką nazębną do wszystkich przestrzeni (im mniej tym lepiej).\
API (%) = suma ocenianych przestrzeni międzyzębowych z płytką / suma wszystkich ocenianych przestrzeni x 100%
*   SBI (Sulcus Bleeding Index) - stosunek przestrzeni międzyzębowych, gdzie przy badaniu wystąpiło krwawienie do wszystkich przestrzeni (im mniej tym lepiej).
*   PI (Plaque Index) - miara obecności płytki nazębnej wokół zęba. Wartość dla danego zęba to suma dla wszystkich powierzchni zęba podzielona przez liczbę powierzchni.

image.png

*   GI (gingival index) - miara stanu dziąsła (od 0 do 3, im mniej tym lepiej). Wartość dla zęba to suma dla wszystkich badanych powierzchni podzielona przez liczbę powierzchni.
image.png

*    PPD (periodontal pocket depth) - głębokość kieszonki przyzębnej, w mm (zdrowy człowiek ma średnio od 0 do 2mm)

*    TWI - miara starcia zębów (od 0 do 4, im mniej tym lepiej)

*     interleukina - wskaźnik występowania stanu zapalnego.

*     wywiad dotyczący oznak bruksizmu - pacjenci byli pytani o występowanie objawów charakterystycznych dla bruksizmu. Komórki przyjmują wartości prawda/fałsz.
